```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# ============================================
# SMART DOCUMENTATION RETRIEVAL ENGINE
# ============================================

# 1. Build TF-IDF index from all documents
corpus = [doc['content'] for doc in documentation]
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)

print(f"✅ Search index built!")
print(f"Total documents indexed: {len(corpus)}")
print(f"Vocabulary size: {len(tfidf_vectorizer.get_feature_names_out())} words")

# 2. Search function
def search_documentation(query, top_k=3):

    # Vectorize the query
    query_vector = tfidf_vectorizer.transform([query])

    # Calculate similarity with all documents
    similarities = cosine_similarity(query_vector, tfidf_matrix).flatten()

    # Get top results
    top_indices = similarities.argsort()[::-1][:top_k]

    print("\n" + "="*60)
    print(f"🔍 Query: '{query}'")
    print("="*60)

    for rank, idx in enumerate(top_indices):
        score = similarities[idx]
        if score > 0:
            doc = documentation[idx]
            print(f"\n📄 Result {rank+1} — Score: {score:.3f}")
            print(f"📁 Category: {doc['category']}")
            print(f"🌐 Language: {doc['language']}")
            print(f"📝 Content: {doc['content']}")
            print("-"*60)

# 3. Test with real questions
print("\n🧪 TESTING THE SEARCH ENGINE:")

search_documentation("What is the oil quantity for Groupe Demarrage engines?")
search_documentation("Quelle est la référence de la courroie pour groupe clim?")
search_documentation("How to change the belt step by step?")
search_documentation("Quelle batterie utiliser pour groupe electrogene?")
```

```python
from IPython.display import display, HTML
import ipywidgets as widgets

# Title
display(HTML("""
<div style="background-color:#1a1a2e; padding:20px;
            border-radius:10px; margin-bottom:10px">
    <h2 style="color:#00d4ff; text-align:center;">
        📄 Smart Documentation Retrieval System
    </h2>
    <p style="color:#ffffff; text-align:center;">
        Ask any question about your engines in English or French
    </p>
</div>
"""))

# Query input
query_input = widgets.Text(
    placeholder='e.g. What oil does ENG007 use? / Quelle courroie pour groupe clim?',
    description='Question:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='600px')
)

# Search button
button = widgets.Button(
    description='🔍 Search Documentation',
    button_style='primary',
    layout=widgets.Layout(width='600px', height='40px')
)

# Output area
output = widgets.Output()

def on_search_click(b):
    with output:
        output.clear_output()

        query = query_input.value.strip()

        if not query:
            display(HTML("<p style='color:red'>⚠️ Please enter a question!</p>"))
            return

        # Search documentation
        query_vector = tfidf_vectorizer.transform([query])
        similarities = cosine_similarity(query_vector, tfidf_matrix).flatten()
        top_indices = similarities.argsort()[::-1][:3]

        display(HTML(f"""
        <div style="background-color:#1a1a2e; padding:10px;
                    border-radius:8px; margin:10px 0">
            <h3 style="color:#00d4ff;">🔍 Results for: "{query}"</h3>
        </div>
        """))

        found = False
        for rank, idx in enumerate(top_indices):
            score = similarities[idx]
            if score > 0:
                found = True
                doc = documentation[idx]

                # Color by category
                cat_colors = {
                    'Engine Specification': '#3498db',
                    'Spare Parts': '#2ecc71',
                    'Coolant Specification': '#1abc9c',
                    'Battery Specification': '#f39c12',
                    'Maintenance Procedure': '#9b59b6',
                    'Safety': '#e74c3c'
                }
                color = cat_colors.get(doc['category'], '#ffffff')
                lang_flag = '🇬🇧' if doc['language'] == 'EN' else '🇫🇷'

                display(HTML(f"""
                <div style="background-color:#16213e; padding:15px;
                            border-radius:8px; margin:8px 0;
                            border-left:4px solid {color}">
                    <p style="color:{color}; margin:0;">
                        <b>📄 Result {rank+1}</b> —
                        {doc['category']} {lang_flag} —
                        Score: {score:.3f}
                    </p>
                    <p style="color:#ffffff; margin:5px 0;">
                        {doc['content']}
                    </p>
                </div>
                """))

        if not found:
            display(HTML("""
            <p style='color:#e74c3c'>
            ❌ No results found. Try different keywords.
            </p>
            """))

button.on_click(on_search_click)

display(query_input)
display(button)
display(output)
```
